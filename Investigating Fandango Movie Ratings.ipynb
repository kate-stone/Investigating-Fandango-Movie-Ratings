# Investigating Fandango Movie Ratings
In 2015 Data Journalist Walt Hickey from FiveThirtyEight published strong evidence to suggest the rating system used by online movie ratings aggregator Fandango was biased and dishonest. Hickey discovered a significant discrepancy between the number of stars displayed to users and the actual rating, which he was able to find in the HTML of the page. Fandango attributed the discrepancy to a bug in their system rather than a diliberate act. They claim the bug has now been fixed

The objective of this project is to analyse more recent movie ratings data to see if the discrepency has truely been fixed. The following datasets, which are publicly available on GitHub, will be used to complete the analysis:
- Movie ratings data Walt Hickey used in his article. This data will be used to analyze the characteristics of Fandango's rating system previous to his analysis. Available [here](https://github.com/fivethirtyeight/data/blob/master/fandango/fandango_scrape.csv);
- Movie ratings data collected by Dataquest for movies released in 2016 and 2017. This data will be used to analyse the rating system's characteristics after Hickey's analysis. Available [here](https://github.com/mircealex/Movie_ratings_2016_17/blob/master/movie_ratings_16_17.csv).

    # Explore the Data

```python
# Read in Walt Hickey Data
import pandas as pd
hickey = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/fandango/fandango_score_comparison.csv")
dq = pd.read_csv("https://raw.githubusercontent.com/mircealex/Movie_ratings_2016_17/master/movie_ratings_16_17.csv")
```

```python
# Explore Hickey
hickey.head()
```

```python
# Explore dq
dq.head()
```

```python
# Extract columns that offer info about Fandango ratings in separate variables
hickey_fandango = hickey[["FILM", "Fandango_Stars", "Fandango_Ratingvalue", "Fandango_votes", 
                          "Fandango_Difference"]].copy()
dq_fandango = dq[["movie", "year", "fandango"]].copy()
hickey_fandango.head()
```

```python
dq_fandango.head()
```

# Compare the Sampling 
Before contining with the analysis, lets compare the two samples to see whether they are representative of the population:
#### Hickey Sample:
The movie ratings for Hickey's analysis were sampled based on the following criteria:

- The movie must have had at least 30 fan ratings on Fandango's website at the time of sampling (24-Aug-2015);
- The movie must have had tickets on sale in 2015.

Straight away it is clear the sampling was not at random as the movie had to have atleast 30 fan ratings. It's also doubtful this sample is representative of the entire population of interest due to the fact it is subject to temporal trends.
#### Dataquest Sample:
Dataquest compiled their sample based on this criteria:

- The movie must have been released in 2016 or later;
- The movie must have had a considerable number of votes and reviews.

This sample is also subject to temporal trends.

# Change the Goal of the Analysis

Due to the discovery that the samples are unlikely to be representative of the population of interest it is unlikely it will be possible to accurately answer the research question. In order to move forward with the analysis, the initial goal will need to be tweaked slightly. 

Instead of finding out whether or not there has been any change in Fandango's rating system after Hickey's analysis, the focus will become investigating if the way Fandango users rate popular movies changed between 2015 and 2016. In other words: Are users rating films higher or lower?

# Confirm Both Samples Contain "Popular" Movies

A new goal has been established and there are two populations of interest that will be compared with each other:

- All Fandango's ratings for popular movies released in 2015;
- All Fandango's ratings for popular movies released in 2016.

Using Hickey's benchmark of 30 fan ratings, a "popular" movie will be defined as one that has atleast 30 fan ratings on Fandango's website. Hickey's sample is already restricted to movies that fit this definition but the representativity of the dataquest sample will need to be checked.

To do this, 10 movies from the dataquest dataset will be randomly sampled and the fan ratings on Fandango's website will be manually checked. Ideally, at least 8 out of the 10 movies will have 30 fan ratings or more.

```python
dq_fandango.sample(10, random_state=1)
```

Manually checking the above movies against Fandango's website revealed 90% are popular. This is enough to move forward with a bit more confidence.

# Isolate the Samples

Taking a close look at the datasets it is evident that some movies were not released in 2015 and 2016. It is necessary to isolate only the sample points that belong to the populations of interest to move forward with the analysis.

```python
data_2015 = hickey_fandango[hickey_fandango["FILM"].str.contains("2015")].copy()
data_2016 = dq_fandango[dq_fandango["year"] == 2016].copy()
```

# Compare Distribution Shapes for 2015 and 2016

The new goal is to determine whether the way users rates films has shifted at all between 2015 and 2016. To start the investigation, a high-level comparison between the shapes of the distributions of movie ratings for both samples will be made.

```python
import matplotlib.pyplot as plt
from numpy import arange
%matplotlib inline
data_2015['Fandango_Stars'].plot.kde(label = '2015', legend = True)
data_2016['fandango'].plot.kde(label = '2016', legend = True)
plt.xlabel("Rating")
plt.ylabel("Density")
plt.style.use('fivethirtyeight')
plt.xticks(arange(0,5.1,.5))
plt.title("Comparing distribution shapes for Fandango's ratings\n(2015 vs 2016)",
          y = 1.07) 
plt.xlabel('Stars')
plt.xlim(0,5) 
plt.show()
```

#### Observations:
- Both distributions are strongly skewed to the left;
- The distribution has shifted slightly in 2016 to the left relative to the 2015 distribution.

It would appear that popular movies received lower ratings in 2016 compared to 2015. 

# Compare Frequency Distribution Tables

The kernel density plots reveal a clear difference between the two distributions. Lets compare the frequency distribution tables to see if they show a similar picture. The samples have a different number of movies so they will be compared using percentages. 

```python
data_2015['Fandango_Stars'].value_counts(normalize=True).sort_index()*100
```

```python
data_2016['fandango'].value_counts(normalize=True).sort_index()*100
```

Looking at the tables it is clear that there are far less movies with ratings of 4.5 and 5 stars in the 2016 distribution compared to the 2015 distribution. The minimum rating is also lower in 2016 â€” 2.5 instead of 3 stars, the minimum of 2015. There clearly is a difference between the two frequency distributions.

However, there were actually more films that received ratings of 3.5 and 4 stars in the 2016 distribution compared to the 2015 distribution. This challenges the direction of the change seen on the kernel density plots.

# Determining the Direction of Change

To determine the direction of change more precisely lets take a look at a few summary statistics.

```python
mean_2015 = data_2015['Fandango_Stars'].mean()
mean_2016 = data_2016['fandango'].mean()

median_2015 = data_2015['Fandango_Stars'].median()
median_2016 = data_2016['fandango'].median()

mode_2015 = data_2015['Fandango_Stars'].mode()[0] # the output of Series.mode() is a bit uncommon
mode_2016 = data_2016['fandango'].mode()[0]

summary = pd.DataFrame()
summary['2015'] = [mean_2015, median_2015, mode_2015]
summary['2016'] = [mean_2016, median_2016, mode_2016]
summary.index = ['mean', 'median', 'mode']
summary
```

```python
plt.style.use('fivethirtyeight')
summary['2015'].plot.bar(color = '#0066FF', align = 'center', label = '2015', width = .25)
summary['2016'].plot.bar(color = '#CC0000', align = 'edge', label = '2016', width = .25,
                         rot = 0, figsize = (8,5))

plt.title('Comparing summary statistics: 2015 vs 2016', y = 1.07)
plt.ylim(0,5.5)
plt.yticks(arange(0,5.1,.5))
plt.ylabel('Stars')
plt.legend(framealpha = 0, loc = 'upper center')
plt.show()
```

Although the median has remained the same for both distributions, the mode and mean have shifted downwards. This confirms the direction of change seen in the kernal density plot.

# Conclusion

The analysis showed there's indeed a slight difference between Fandango's ratings for popular movies in 2015 and Fandango's ratings for popular movies in 2016. It was also determined that, on average, popular movies released in 2016 were rated lower on Fandango than popular movies released in 2015.

It isn't possible to be completely sure what caused the change, but there's a very high chance it was the result of Fandango fixing the biased rating system after Hickey's analysis.
